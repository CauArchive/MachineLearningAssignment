# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## plot the input data

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    # features: 2
    distance = np.sum((feature-center)**2, axis=1)
    distance = np.sqrt(distance)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    same_label = (label_feature == label)
    parsed_feature = feature[same_label]
    if(np.size(parsed_feature, 0) == 0):
        return None

    centroid = np.mean(parsed_feature, axis=0)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    label_feature = np.argmin(distance, axis=1)
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

```python
def compute_loss(distance, label_feature, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    parsed_distance = distance[np.arange(len(distance)), label_feature]
    loss_sum = np.power(parsed_distance, 2)
    loss = np.sum(loss_sum) / len(feature)
  
    return loss
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 100     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#

for i in range(number_iteration):
      # shape [1000, 2]
      distance_iteration_02 = np.zeros((number_data, number_cluster))
      for j in range(number_cluster):
            cur_centroid = compute_centroid(feature, label_feature_02, j)
            if(cur_centroid is None):
                  centroid_iteration_02[i, j] = centroid_iteration_02[i-1, j]
            else:
                  centroid_iteration_02[i, j] = cur_centroid
            distance_iteration_02[:, j] = compute_distance(feature, centroid_iteration_02[i, j])
      label_feature_02 = compute_label(distance_iteration_02)
      loss_iteration_02[i] = compute_loss(distance_iteration_02, label_feature_02, feature)
```
