## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    feature = np.ones((point.shape[0], 20))
    for i in range(feature.shape[0]):
        feature[i, 1] = point[i, 0]
        feature[i, 2] = point[i, 1]
        feature[i, 3] = point[i, 0]**2
        feature[i, 4] = point[i, 0] * point[i, 1]
        feature[i, 5] = point[i, 1]**2
        feature[i, 6] = point[i, 0]**3
        feature[i, 7] = point[i, 0]**2 * point[i, 1]
        feature[i, 8] = point[i, 0] * point[i, 1]**2
        feature[i, 9] = point[i, 1]**3
        feature[i, 10] = np.cos(point[i, 0])
        feature[i, 11] = np.sin(point[i, 0])
        feature[i, 12] = np.cos(point[i, 1])
        feature[i, 13] = np.sin(point[i, 1])
        feature[i, 14] = np.cos(point[i, 0]**2)
        feature[i, 15] = np.sin(point[i, 0]**2)
        feature[i, 16] = np.cos(point[i, 1]**2)
        feature[i, 17] = np.sin(point[i, 1]**2)
        feature[i, 18] = np.cos(point[i, 0] * point[i, 1])
        feature[i, 19] = np.sin(point[i, 0] * point[i, 1])

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = theta @ feature.T

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = 1 / (1 + np.exp(-z))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = sigmoid(compute_linear_regression(theta, feature))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = compute_logistic_regression(theta, feature)
    residual = -(label * np.log1p(value) + (1 - label) * np.log1p(1 - value))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = compute_residual(theta, feature, label)
    number_data = value.shape[0]
    loss = (np.sum(value) / number_data) + (alpha * np.sum(theta.T @ theta) / 2)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = compute_logistic_regression(theta, feature)
    number_data = value.shape[0]
    gradient = ((value - label) @ feature / number_data) + (alpha * theta)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = compute_logistic_regression(theta, feature)
    number_data = value.shape[0]
    accuracy = 0
    for i in range(number_data):
        if (value[i] >= 0.5 and label[i] == 1) or (value[i] < 0.5 and label[i] == 0):
            accuracy += 1
    accuracy = accuracy / number_data

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
number_iteration    = 2000 # you can change this value as you want 
learning_rate       = 0.1 # you can change this value as you want 
number_feature      = 20 # you can change this value as you want
alpha               = 0.0001 # you can change this value as you want

theta                       = np.zeros(number_feature)
loss_iteration_train        = np.zeros(number_iteration)
loss_iteration_test         = np.zeros(number_iteration)
accuracy_iteration_train    = np.zeros(number_iteration)
accuracy_iteration_test     = np.zeros(number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
computed_feature_train      = compute_feature(data_train_point)
computed_feature_test       = compute_feature(data_test_point)

for i in range(number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
 
    theta           = theta - learning_rate * compute_gradient(theta, computed_feature_train, data_train_label, alpha)
    loss_train      = compute_loss(theta, computed_feature_train, data_train_label, alpha)
    loss_test       = compute_loss(theta, computed_feature_test, data_test_label, alpha)
    accuracy_train  = compute_accuracy(theta, computed_feature_train, data_train_label)
    accuracy_test   = compute_accuracy(theta, computed_feature_test, data_test_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    loss_iteration_train[i]     = loss_train
    loss_iteration_test[i]      = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i]  = accuracy_test

theta_optimal = theta
```
